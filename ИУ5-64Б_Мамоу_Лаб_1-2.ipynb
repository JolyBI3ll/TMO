##**О наборе данных**

Давайте разберем предоставленный набор данных и объясним каждый столбец вместе с его категориальными значениями:


1. rest_bp: Артериальное давление в состоянии покоя (мм рт.ст.)

  - Числовая непрерывная переменная, представляющая артериальное давление пациента в состоянии покоя.

2. chest_pain: Тип боли в груди

  -  Категориальная переменная со следующими значениями:
    - 0: Типичная стенокардия
    - 1: Атипичная стенокардия
    - 2: Боль, не связанная с ангиной
    - 3: Бессимптомная

3. thalassemia: Тип талассемии

  - Категориальная переменная со следующими значениями:
    - 0: Нормальная
    - 1: Исправленный дефект
    - 2: Возраст обратимого дефекта

4. agа (годы)

  - Числовая непрерывная переменная, представляющая возраст пациента.

5. fasting_bs: Уровень сахара в крови натощак > 120 мг/дл

  - Категориальная переменная со следующими значениями:
    - 0: Уровень сахара в крови натощак <= 120 мг/дл
    - 1: Уровень сахара в крови натощак > 120 мг/дл

6. max_hr: Достигнутая максимальная частота сердечных сокращений (ударов в минуту)

  - Числовая непрерывная переменная, представляющая максимальную частоту сердечных сокращений, достигнутую во время тренировки.

7. exercise_angina: Стенокардия, вызванная физической нагрузкой

  - Категориальная переменная со следующими значениями:
    - 0: Стенокардия, вызванная физической нагрузкой, отсутствует
    - 1: Стенокардия, вызванная  физической нагрузкой

8. Gender: Пол пациента

  - Категориальная переменная со следующими значениями:
    - 0: Женский пол
    - 1: Мужской

9. st_slope: наклон сегмента ST

  - Категориальная переменная со следующими значениями:
    - 0: Восходящий
    - 1: Плоский
    - 2: Нисходящий

10. cholesterol: Уровень холестерина в сыворотке крови (мг/дл)

    - Числовая непрерывная переменная, представляющая уровень холестерина в сыворотке крови пациента.

11. st_depression: депрессия ST, вызванная физической нагрузкой относительно покоя

  - Числовая непрерывная переменная, представляющая депрессию ST, вызванную физической нагрузкой относительно покоя.

12. rest_ecg: Результаты электрокардиографии в состоянии покоя
    - Категориальная переменная со следующими значениями:
      - 0: Нормальный
      - 1: Аномалия зубца ST-T
      - 2: Вероятная или определенная гипертрофия левого желудочка
13. num_vessels: Количество крупных сосудов, окрашенных при рентгеноскопии

  - Категориальная переменная со следующими значениями:
      - 0: Нет
      - 1: Один
      - 2: Два
      - 3: Три

14. diagnosis: Диагноз болезни сердца
  - Категориальная переменная со следующими значениями:
      - 0: Отсутствие болезни сердца (отрицательный диагноз)
      - 1: Наличие болезни сердца (положительный диагноз)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('heart_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# List of columns with numeric data
numeric_features = ['age', 'rest_bp', 'cholesterol', 'max_hr', 'st_depression', 'diagnosis']

# List of columns with categorical data
cat_features = ['gender', 'chest_pain', 'fasting_bs', 'rest_ecg', 'exercise_angina', 'st_slope', 'num_vessels', 'thalassemia']

# Print the lists of features
print("Numeric features:", numeric_features)
print("Categorical features:", cat_features)
```

```python
df[cat_features] = df[cat_features].astype(object)

df.dtypes
```

## categorical data analysis

```python
numeric_df = df[numeric_features]

# Calculate the correlation between selected columns
corr_data = numeric_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_data, annot=True, cmap='Reds', linewidths=0.1)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

## Numerical data Analysis

```python
age_col = df['age']

# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Generate a histogram for the 'age' column
plt.hist(age_col, bins=20, color='skyblue', alpha=0.7)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
trestbps_col = df['rest_bp']

plt.hist(trestbps_col, bins=20, color='lightcoral', alpha=0.7)

plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Frequency')
plt.title('Resting Blood Pressure Distribution')
plt.show()
```

```python
chol_col = df['cholesterol']
plt.hist(chol_col, bins=20, color='lightgreen', alpha=0.7)

plt.xlabel('Cholesterol (mg/dl)')
plt.ylabel('Frequency')
plt.title('Cholesterol Distribution')
plt.show()
```

```python
thalach_col = df['max_hr']
plt.hist(thalach_col, bins=20, color='lightblue', alpha=0.7)

plt.xlabel('Maximum Heart Rate (beats per minute)')
plt.ylabel('Frequency')
plt.title('Maximum Heart Rate Distribution')
plt.show()
```

```python
oldpeak_col = df['st_depression']

plt.hist(oldpeak_col, bins=20, color='lightblue', alpha=0.7)
plt.xlabel('ST Depression')
plt.ylabel('Frequency')
plt.title('Distribution of ST Depression')
plt.show()
```
